```python
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt

img = mpimg.imread('dog.jpeg')
imgplot = plt.imshow(img)
plt.show()
```

# Q12: Lightness method

```python
gray_img_01 = (np.max(img, axis=-1) + np.min(img, axis=-1)) / 2.0

# Display grayscale value at top-left pixel
gray_img_01[0, 0]
```

# Q13: Average method

```python
gray_img_02 = np.mean(img, axis=-1)

# Display grayscale value at top-left pixel
print(gray_img_02[0, 0])
```

# Q14: Luminosity method

```python
gray_img_03 = 0.21 * img[..., 0] + 0.72 * img[..., 1] + 0.07 * img[..., 2]

# Display grayscale value at top-left pixel
print(gray_img_03[0, 0])
```
